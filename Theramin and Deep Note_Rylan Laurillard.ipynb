### 1. Sine wave at 220hz for 1 second

```python
import numpy as np
import IPython.display as ipd

def sinusoid(freq = 440.0, dur = 1.0, srate = 44100.0, amp = 1.0, phase = 0.0, glide = 1.0):
    t = np.linspace(0,dur,srate*dur)
    data =  amp * np.sin(2*np.pi*freq*(t**glide)+phase)
    return data

srate = 44100
dur = 1
freq = 440
amp = 1.0
phase = 0.0
glide = 1.0

data = sinusoid(freq, dur, srate, amp, phase, glide)
ipd.Audio(data,rate=srate)
```

### 3-4 Additive intrument, Control by Hz/Note

Takes a letter input from c-b and will return the corresponding note in the c major scale. Alternatively, takes an integer/float value and plays said note (in hertz).

```python
def instrument(freq = 440, dur = 1, srate = 44100, amp = 1.0, phase = 0.0, glide = 1.0):
    inst1 = sinusoid(freq, dur, srate, amp, phase, glide)
    inst2 = sinusoid(freq*2, dur, srate, amp*0.8, phase, glide)
    inst3 = sinusoid(freq*3, dur, srate, amp*0.6, phase, glide)
    inst4 = sinusoid(freq*4, dur, srate, amp*0.4, phase, glide)

    inst = np.vstack([inst1, inst2, inst3, inst4])
    return inst

def get_note():
    y = True
    while y == True:
        input_note = input("Enter a note (lowercase letter or # of Hz): ")
        try:
            if input_note == 'c':
                y = False
                return 261.63
            elif input_note == 'd':
                y = False
                return 293.66
            elif input_note == 'e':
                y = False
                return 329.63
            elif input_note == 'f':
                y = False
                return 349.23
            elif input_note == 'g':
                y = False
                return 392.00
            elif input_note == 'a':
                y = False
                return 440.00
            elif input_note == 'b':
                y = False
                return 493.88
            else:
                input_note = float(input_note);
                y = False
                return input_note
        except:
            print("Wrong input, please try again.")

note = get_note()
inst = instrument(freq = note, dur = 2)
ipd.Audio(inst,rate=srate)
```

### 5. ADSR Envelope

```python
def adsr(note, attack = 0.0, decay = 0.0, sustain = 1.0, release = 0.0, srate = 44100):
    dur = (len(note) + release*srate)
    
    #attack - amplitude goes from 0-1.0 in time = 0-(attack)
    t = np.linspace(0,attack,srate*attack)
    note_attack = t * note
    
    #decay - amplitude goes from 1.0-sustain in time = (attack)-(decay)
    t = np.linspace(attack,decay,srate*decay)
    note_decay = t * note
    
    #sustain - amplitude =(sustain) from time = (decay)-len(note)
    t = np.linspace(decay,sustain,srate*len(note))
    note_sustain = t * note
    
    #release - amplitude goes from (sustain)-0.0 in time = len(note)-(release)
    t = np.linspace(len(note),len(note)+release,srate*release)
    note_release = t * note
    
    note = np.hstack([note_attack, note_decay, note_sustain, note_release])
    
    return note
    

note = sinusoid(freq = 440, dur = 1, srate = 44100, amp = 1.0, phase = 0.0, glide = 1.0)

filtered_note = adsr(note)
#ipd.Audio(filtered_inst, rate = srate)
```
