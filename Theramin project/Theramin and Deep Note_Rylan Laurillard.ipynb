### 1. Sine wave at 220hz for 1 second

```python
import numpy as np
import IPython.display as ipd

def sinusoid(freq = 440.0, dur = 1.0, srate = 44100.0, amp = 1.0, phase = 0.0, glide = 1.0):
    t = np.linspace(0,dur,srate*dur)
    data =  amp * np.sin(2*np.pi*freq*(t**glide)+phase)
    return data

srate = 44100
dur = 1
freq = 440
amp = 1.0
phase = 0.0
glide = 1.0

data = sinusoid(freq, dur, srate, amp, phase, glide)
ipd.Audio(data,rate=srate)
```

### 3-4 Additive intrument, Control by Hz/Note

Takes a letter input from c-b and will return the corresponding note in the c major scale. Alternatively, takes an integer/float value and plays said note (in hertz).

```python
def instrument(freq = 440, dur = 1, srate = 44100, amp = 1.0, phase = 0.0, glide = 1.0):
    inst1 = sinusoid(freq, dur, srate, amp, phase, glide)
    inst2 = sinusoid(freq*2, dur, srate, amp*0.8, phase, glide)
    inst3 = sinusoid(freq*3, dur, srate, amp*0.6, phase, glide)
    inst4 = sinusoid(freq*4, dur, srate, amp*0.4, phase, glide)

    inst = np.vstack([inst1, inst2, inst3, inst4])
    return inst

def get_note():
    y = True
    while y == True:
        input_note = input("Enter a note (lowercase letter or # of Hz): ")
        try:
            if input_note == 'c':
                y = False
                return 261.63
            elif input_note == 'd':
                y = False
                return 293.66
            elif input_note == 'e':
                y = False
                return 329.63
            elif input_note == 'f':
                y = False
                return 349.23
            elif input_note == 'g':
                y = False
                return 392.00
            elif input_note == 'a':
                y = False
                return 440.00
            elif input_note == 'b':
                y = False
                return 493.88
            else:
                input_note = float(input_note);
                y = False
                return input_note
        except:
            print("Wrong input, please try again.")

note = get_note()
inst = instrument(freq = note, dur = 2)
ipd.Audio(inst,rate=srate)
```

### 5. ADSR Envelope

```python
import matplotlib.pyplot as plt

def adsr(note, attack = 1.0, decay = 1.0, sustain = 0.5, release = 1.0, srate = 44100):
    dur = (len(note) + release*srate)
    note_attack =[]
    note_decay = []
    note_sustain = []
    note_release = []
    
    #attack - amplitude goes from 0-1.0 in time = 0-(attack)
    if attack != 0:
        t = np.linspace(0, attack, int(srate*attack))
        counter = 1
        
        for n in note:
            if counter > srate*attack: 
                break
            note_attack.append(n)
            counter+=1
            
        note_attack = np.multiply(note_attack, t)

    #decay - amplitude goes from 1.0-sustain in time = (attack)-(decay)
    if decay != 0:
        t = np.linspace(1.0, sustain, int(srate*decay))
        counter = 1
        
        for n in note:
            if counter <= (srate*attack):
                counter +=1
            elif counter > ((srate*attack)+(srate*decay)): 
                break
            else:
                note_decay.append(n)
                counter +=1
        
        note_decay = np.multiply(note_decay, t)
     
    #sustain - amplitude = (sustain) from time = (attack + decay)-(len(note))
    if sustain != 0:
        t = np.linspace(sustain, sustain, int(len(note)-((srate*attack)+(srate*decay))))
        counter = 1
        
        for n in note:
            if counter <= ((srate*attack)+(srate*decay)):
                counter +=1
            else:
                note_sustain.append(n)
                counter +=1
        
        note_sustain = np.multiply(note_sustain, t)
    
    #release - amplitude goes from (sustain)-0.0 in time = (len(note))-(release)
    if release != 0:
        t = np.linspace(sustain, 0, int(srate*release))
        counter = 1
  
        for n in note:
            if counter > (srate*release):
                break
            else:
                note_release.append(n)
                counter +=1
        
        note_release = np.multiply(note_release, t)
    
    note = np.hstack([note_attack, note_decay, note_sustain, note_release])
    return note
    

note = sinusoid(freq = 440, dur = 5, srate = 44100, amp = 1.0, phase = 0.0, glide = 1.0)


filtered_note = adsr(note, 1.0, 0.9, 0.5, 1.0)
#print(len(filtered_note))

x = np.linspace(0, 2 * np.pi, 264600)
plt.plot(x, filtered_note)
plt.grid(True)
plt.xlim(0, 10)
plt.show()

ipd.Audio(filtered_note, rate = srate)
```
